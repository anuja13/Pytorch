```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = torch.arange(0,500,6.5)
noise = torch.randn(len(x_data))
y_data = 2*x_data + 100*noise
```

```python
plt.scatter(x_data.numpy(),y_data.numpy())
plt.ylabel('Y')
plt.xlabel('X')
plt.title('Data Distribution')
```
